# Functions 



```python
def mae(y, ypred):    
    return np.sum([abs(y[i]-ypred[i]) for i in range(len(y))]) / len(y) 

import matplotlib.pyplot as plt


def make_bar_plot(df, title):
    
    plt.figure(figsize = [10,10])
    df.plot(kind='barh', color = 'steelblue')
    plt.title(title, size=14, fontdict={'fontweight': 'bold'})
    plt.show(title)

    
def make_pred_plot(y_test, y_pred, lim, title='real versus predicted bp', xlabel='real_bp', ylabel='predicted_bp'):

    plt.figure(figsize = [10,10])
    plt.scatter(y_test, y_pred, marker='.')
    ax = plt.gca()
    ax.set_xlabel(xlabel, size=14)
    ax.set_ylabel(ylabel, size=14)
    
    x = np.linspace(*ax.get_xlim())
    ax.plot(x, x, 'g--')
    if lim is not None:
      ax.set_xlim(lim)
      ax.set_ylim(lim)

    plt.title(title, size=14, fontdict={'fontweight': 'bold'})
    plt.show()
    

def get_feature_importance(CV, features):
    
    try:
       feat_imp = CV.best_estimator_.named_steps.reg.feature_importances_
    except:
       feat_imp = CV.best_estimator_.named_steps.reg.coef_

    selected_features = features[:]
    df_feat_imp = pd.DataFrame(feat_imp, index = selected_features)
    make_bar_plot(df_feat_imp[0].sort_values(), title = 'feature importances')
    
    return df_feat_imp, selected_features
```

# Preprocessing

```python
df = df.rename(columns={'icustay_id' :'id'})
               
df = df.drop_duplicates(subset = ['icustay_id'], keep = 'first')

df_model = df[df['invasive_value'].between(40, 140)]
df_model = df_model[df_model['ninv_value'].between(40, 140)]

df_model['diff'] = df_model['invasive_value'] - df_model['ninv_value']
df_model['perc_difference'] = df_model['diff'] / df_model['ninv_value']


df_model = df_model[df_model['femoral_yn'] ==1]
df_model['vasopressor'] = df_model['vasopressor'].fillna(0)
```

# Select outcome feature and features to use

```python
outcome_feature = 'diff' #difference invasive_value

input_features = df_model.columns.tolist()
del_columns = ['id', 'time_diff', 'invasive_time', 'ninv_time',
               'ranky', 'binary_flag', 'difference', 'subject_id', 'admission_id', 'subject_id',
              'expire_flag','last_wardid','icu_length_of_stay_d', 'icustay_id', 'hadm_id', 'length_o_stay_d_precise',
              'first_wardid', 'invasive_value','difference', 'perc_difference', 'femoral_yn', 'diff']

columns_to_use = list(set(input_features) - set(del_columns))

columns_to_use = [x for x in columns_to_use if df_model[x].dtype != 'O']
```

```python
X = df_model[columns_to_use].values
y = df_model[outcome_feature].values
features = columns_to_use[:]
```

# Train with grid search and CV

```python

param_grid_lgbm = {
    'reg__n_estimators':[200],
    'reg__learning_rate': [0.1],
    'reg__num_leaves': [25, 50, 100],
    'reg__min_data_in_leaf': [5, 10, 20],
    #'feat_select__k': ['all'],
} 


param_grid_rf = {
    'reg__n_estimators':[500],
    'reg__min_samples_leaf': [3, 5, 10]
} 


param_grid_lr = {
} 
```

```python
modeltype = 'RF' # LGBM LogisticRegression # RF

if modeltype == 'LinearRegression':
  regressor = LR()
  param_grid = param_grid_lr

elif modeltype == 'LGBM':
  regressor = LGBM()
  param_grid = param_grid_lgbm
  
elif modeltype == 'RF':
  regressor = RFR()
  param_grid = param_grid_rf


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

pipeline = Pipeline(
                    [
                     ('imp', SimpleImputer(strategy = 'median')),
                     ('scaler', StandardScaler()),
                     #('feat_select', SelectKBest(k=5)),
                     ('reg', regressor

                     )
                     
])

CV = GridSearchCV(pipeline, param_grid, cv=3, n_jobs=-1, scoring='r2')
CV.fit(X_train, y_train)   
```

```python
y_pred = CV.predict(X_test)
df_feat_imp, features_selected = get_feature_importance(CV, features)

print('Best_params: %s' % CV.best_params_)
print()
print('Best_score: %s' % np.round(CV.best_score_, 3))
print()
print('MAE on validation set: %s' % (round(mae(y_test, y_pred), 3)))
print()
print ("R-squared on validation set: %.2f" %CV.score(X_test, y_test))

make_pred_plot(y_test, y_pred, lim = [30, 150])
```

```python
# !pip install shap
pipeline = CV.best_estimator_

imp = pipeline.named_steps.imp
scaler = pipeline.named_steps.scaler
X_model = df_model[columns_to_use]
X_imputed = imp.transform(X_model)
X_shap = scaler.transform(X_imputed)
```
